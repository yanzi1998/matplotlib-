```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
```

```python
df = pd.read_excel('体温.xls')
```

```python
x = df['日期']
y = df['体温']
plt.plot(x,y)
```

## 颜色设置


设置值|说明|设置值|说明
---|:--:|---:|:--:
b|蓝色|m|洋红色
g|绿色|y|黄色
r|红色|k|黑色
c|蓝绿色|w|白色
#FFFF00|黄色，十六进制颜色值
0.5|灰度值字符串

```python
plt.plot(x,y,color = 'k',linestyle = '-',marker = 'D')
```
